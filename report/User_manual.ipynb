# 2. Design own configure file


The configure file is composed by key value pairs, and each pairs occupy a line.

$\textit{key = value}$

In order to run the agent with custom configure file, parse the commend below in terminal.

$\textit{python aixi.py -v file.conf }$

In order to swap between different environment, parse the commend below in terminal. The s short for second environment, and can be used as -s< second enviroment time >, time to change

$\textit{python aixi.py -v -s extended_tiger,200 conf/kuhn_poker.conf }$

There options for configure files are below.

- **ct-depth:** maximum depth of the context tree used for prediction.
- **agent-horizon:** the number of percept/action pairs to look forward.    
- **exploration:** probability of playing a random move.
- **explore-decay:** a value between 0.0 and 1.0 that defines the geometric decay
of the exploration rate.
- **terminate-age:** how many agent/environment cycles before the agent needs to finish.
- **mc-simulations:** the number of MC simulations per cycle.
- **environment:** the environment the agent is to interact with.
- **learning-period:** the cycle count.
- **second:**  the second enviroment to run, you may need to swap between environments.Please follow the formats below to specify the parameter <br> -s <name of second enviroment,time to run second environment>




# 3. Experimental Script 

We provide the user a script to running the experiments, and the $\textit{tee}$ commend need to be installed. 

----------------------------

### 3.1 Parameter Options
- **experimental_result:** The folder to store the experimental result. The script will create it if there doesn't existes one. And, all information is stored as log file.  
- **conf_director:** The folder contains the different configure files. Only the file endup with ".conf" will be experimented.
- **performance_inrease_graph:** Reading the specifed log file, and generate a single graph about reward. The format for setting the parameter is <br> < log file >$\text{~}$<'title of the graph'>  
- **interval_performance_inrease_graph:** Reading the specifed log file, and generate a single smoothed graph about the average reward. The interval can be set by using interval parameter. And, the format for setting the parameter is <br>  < log file >$\text{~}$<'title of the graph'>
- **compare_performance_graph:** By supplying the director, it will reading a list of log file. Then, generate the graph for them. The format to giving the value is <br>  < where store the log files >$\text{~}$<"['name of labels',]">$\text{~}$<'title of graph'>. <br>  Be careful, the order for the name of labels need to corresponds with the alphabeta order of log files.
- **custom_name:** The log file for different configure files wil be stored as the name of configure file plus the custom_name. if no custom name is supplied, the current time will be considered as custom name. 
- **interval:** In order to smooth the history reward, the interval need to be supplied. The default value will be 50.
- **type of reward:** Used to indicate whether we should read average reward or reward from the log file.


----------------------------

### 3.2 Running experiments

Here is a example to run the experiments.

$\textit{python experimental_script.py  -c conf_director -e experimental_result }$

----------------------------

### 3.3 Generating graphs

Here is two example to generate the graph, and the demonstration of graph style.

$\textit{python experimental_script.py  -v experimental_result~"['coin flip','coin flip compare']"~'coin flip' -g 10}$
